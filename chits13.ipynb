```python
#Step 1: Import Required Libraries
import pandas as pd
import numpy as np
```

```python
#Step 2: Load the Dataset Properly
# Load dataset
df = pd.read_csv('AirQualityUCI.csv')

# Display first 5 rows
df.head()
```

```python
#Step 3: Check Dataset Information
# Shape of dataset
print("Shape of Dataset:", df.shape)

# Column names
print("\nColumn Names:")
print(df.columns)

# Dataset information
print("\nDataset Information:")
df.info()
```

```python
#Step 4: Check Missing Values
# Check missing values
print(df.isnull().sum())
```

```python
#a DATA CLEANING
#Step 5: Remove Duplicate Rows
# Count duplicate rows
duplicates = df.duplicated().sum()

print("Duplicate Rows:", duplicates)

# Remove duplicates
df = df.drop_duplicates()

# Check new shape
print("New Shape:", df.shape)
```

```python
#Step 6: Handle Missing Values
# Replace -200 with NaN
df.replace(-200, np.nan, inplace=True)
print(df.isnull().sum())
```

```python
# Step 7: Convert Date and Time into Datetime Format

# Combine Date and Time columns
if 'Date' in df.columns and 'Time' in df.columns:
    
    df['Datetime'] = pd.to_datetime(
        df['Date'] + ' ' + df['Time'],
        dayfirst=True
    )

# Display updated dataset
df.head()
```

```python
#Step 8: Statistical Summary
# Statistical summary
df.describe()
```

```python
# b DATA INTEGRATION
# Step 9: Create Environmental Quality Index
# Create Environmental Index
if {'Temp', 'RH', 'AH'}.issubset(df.columns):
    
    df['Environmental_Index'] = (
        df['Temp'] + df['RH'] + df['AH']
    ) / 3

# Display data
df.head()
```

```python
#Step 10: Create Integrated Dataset
# Select important columns
selected_columns = []

for col in ['Datetime', 'CO(GT)', 'NMHC(GT)', 'NO2(GT)',
            'Temp', 'RH', 'AH', 'Environmental_Index']:
    
    if col in df.columns:
        selected_columns.append(col)

# Create integrated dataset
integrated_df = df[selected_columns]

# Display integrated dataset
print(integrated_df.head())
```

```python
#Step 11: Save Cleaned and Integrated Dataset
# Save final dataset
integrated_df.to_csv('Cleaned_Integrated_AirQuality.csv', index=False)

print("Dataset saved successfully!")
```
