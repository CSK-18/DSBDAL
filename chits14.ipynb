```python
import pandas as pd
import numpy as np
```

```python
#Step 2: Load the Dataset Properly
# Load dataset
df = pd.read_csv('dataset_Facebook.csv')

# Display first 5 rows
df.head()
```

```python
#Step 3: Create Column Subset
subset_columns = df[[
    'pagetotallikes',
    'type',
    'category',
    'postmonth',
    'postweekday',
    'lifetimeposttotalreach'
]]

subset_columns.head()
```

```python
#Step 4: Create Row Subset
high_reach = df[df['lifetimeposttotalreach'] > 50000]

high_reach.head()
```

```python
#Step 5: Create Type-Based Subset
photo_posts = df[df['type'] == 'Photo']

photo_posts.head()
```

```python
# B Merge Data
#Step 6: Create Two DataFrames
df1 = df[['type', 'category', 'postmonth']]

df2 = df[['like', 'share', 'comment']]
```

```python
#Step 7: Merge DataFrames
merged_df = pd.concat([df1, df2], axis=1)

merged_df.head()
```

```python
merged_df.to_csv('Merged_Facebook_Data.csv', index=False)

print("Dataset saved successfully!")
```
